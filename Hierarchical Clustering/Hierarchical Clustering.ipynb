# Hierarchical Clustering

#### Important parameters to consider:
<li> Distance metrics: Euclidian, Manhattan, Cosine

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

#### Creating custom datasets

```python
from sklearn.datasets import make_blobs
```

```python
exampl_data = make_blobs(n_samples = 300, n_features=2, centers=3, cluster_std = 1.3, random_state=42)
```

```python
exampl_data
```

```python
dots = exampl_data[0]
```

```python
dots
```

### Dendograms

```python
import scipy.cluster.hierarchy as sch
```

```python
dendogram_example = sch.dendrogram(sch.linkage(dots, method='ward'))
```

#### Scatterplot

```python
plt.figure(figsize=(10, 7))
plt.scatter(dots[:, 0], dots[:, 1])
plt.show()
```

### Modelling

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hierarchical_clustering = AgglomerativeClustering(n_clusters = 5, affinity= 'euclidean', linkage = 'ward')
```

```python
y_hier_clus = hierarchical_clustering.fit_predict(dots)
```

```python
y_hier_clus
```

```python
plt.figure(figsize=(10,7))
plt.scatter(dots[y_hier_clus == 0, 0], dots[y_hier_clus == 0, 1], c = 'blue')
plt.scatter(dots[y_hier_clus == 1, 0], dots[y_hier_clus == 1, 1], c = 'black')
plt.scatter(dots[y_hier_clus == 2, 0], dots[y_hier_clus == 2, 1], c = 'orange')
plt.scatter(dots[y_hier_clus == 3, 0], dots[y_hier_clus == 3, 1], c = 'pink')
plt.scatter(dots[y_hier_clus == 4, 0], dots[y_hier_clus == 4, 1], c = 'yellow')
plt.show()
```

#### Here we created a dataset with 3 centers and we asked the algorithm to cluster it into 5.
