### <center>CREATING ROC CURVE WITH AUC IN ONE LINE OF CODE

```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import plot_roc_curve
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
```

## BINARY CLASSIFICATION

### Lets load the titanic dataset from seaborn library for the purpose.

```python
data = sns.load_dataset('titanic')
data.head()
```

```python
x = data[['pclass', 'sibsp', 'fare']].values
y = data[['survived']].values
```

```python
x = StandardScaler().fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
model = RandomForestClassifier(random_state=0)
```

```python
model.fit(x_train, y_train)
```

```python
plot_roc_curve(model, x_test, y_test);
```

#### Creating an object named 'disp' for plotting multiple graphs together.

```python
disp = plot_roc_curve(model, x_test, y_test)
```

```python
Logistic_Regression_model = LogisticRegression()
DecisionTreeClassifier_model = DecisionTreeClassifier()
```

```python
Logistic_Regression_model.fit(x_train, y_train)
```

```python
DecisionTreeClassifier_model.fit(x_train, y_train)
```

```python
disp= plot_roc_curve(model, x_test, y_test);
plot_roc_curve(DecisionTreeClassifier_model, x_test, y_test, ax=disp.ax_);
plot_roc_curve(Logistic_Regression_model, x_test, y_test, ax=disp.ax_);
```
